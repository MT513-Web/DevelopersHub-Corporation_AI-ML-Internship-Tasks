## Task 1: Exploring and Visualizing the Iris Dataset

Objective: The goal of this task is to perform an Initial Data Analysis (IDA) on the Iris dataset. 
This involves loading the data, inspecting its structure, calculating descriptive statistics, and creating visualizations to identify patterns and outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
print("Shape of dataset:", df.shape)  
print("Column Names:", df.columns) 

print(df.head())
print("\nDataset Information:")
df.info()
```

### 1. Data Loading and Inspection
* In this initial step, I used Pandas to load the dataset and Seaborn to fetch the built-in Iris data. 
* .head(): Displays the first 5 rows to understand the feature set.
* .info(): Confirms that there are 150 entries with no missing (null) values.

To understand the distribution and scale of our data, I used the .describe() function.
 This provides a detailed statistical overview, including the mean, standard deviation, and range for each numerical feature in the Iris dataset. 
 This step is crucial for identifying the central tendency of the measurements.

```python
print("Summary Statistics:")
print(df.describe())
```

Visualizing the data helps in identifying patterns and clusters that might not be obvious from the numbers alone.
Here, I am creating a "Scatter Plot" to examine the relationship between Sepal Length and Sepal Width. 
By using different colors for each species, we can see how the flowers naturally group together based on their physical attributes.

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width')
plt.show()
```

Observation:The scatter plot shows that the Setosa species forms a very distinct cluster compared to Versicolor and Virginica, making it the easiest species to classify based on sepal measurements.

Distribution of Features: This Histogram helps us observe the frequency distribution of Sepal Length across the dataset. The KDE (Kernel Density Estimate) line provides a smooth estimate of the data distribution.

```python

plt.figure(figsize=(8, 5))
sns.histplot(df['sepal_length'], kde=True, color='purple')
plt.title('Distribution of Sepal Length')
plt.show()
```

Outlier Detection: I used Box Plots to visualize the spread of each feature and identify any potential outliers. This is essential for ensuring data quality before any further analysis or modeling.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette="Set2")
plt.title('Outlier Detection: Box Plot of Iris Features')
plt.ylabel('Measurements (cm)')
plt.show()
```

Key Observations from Task 1:
* The scatter plot clearly shows that different Iris species (especially Setosa) form distinct clusters.
* The summary statistics indicate that there are no missing values in the 150 entries. 
* The box plot helps us verify if any feature has unusual measurements outside the normal range.
